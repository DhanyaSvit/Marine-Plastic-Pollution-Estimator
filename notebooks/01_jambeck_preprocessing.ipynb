##### Calculating Mismanaged waste index
Share of plastic inadequately managed = 45% . 
This column is already our Mismanaged Waste Index (in percentage form).

```python
data["Mismanaged_Waste_Index"] = (data["Share of plastic inadequately managed"] / 100)
```

```python
data["Mismanaged_Waste_Index"].describe()
```

##### Dropping the Unnecessary Columns

```python
data.drop([
    "Year",
    "Per capita waste generation rate",
    "Share of plastics in waste stream",
    "Share of plastic inadequately managed",
    "Municipal waste generated",
    "Plastic waste generated",
    "Inadequately managed plastic waste",
    "Plastic waste littered",
    "Per capita mismanaged plastic waste",
    "Total mismanaged plastic waste in 2025",
    "Per capita plastic waste (kg/person/day)"
], axis=1, inplace=True)
```

#### New Csv file to store cleaned/necessary data
The cleaned dataset is saved as a new CSV file named jambeck_cleaned.csv for further analysis and model development.

```python
data = data[[
    "Entity",
    "Coastal population",
    "Total mismanaged plastic waste in 2010",
    "Mismanaged_Waste_Index"
]]
```

#### Skewness Analysis
Skewness was calculated to evaluate the distribution shape of key numerical variables.

```python
data["Total mismanaged plastic waste in 2010"].skew()
```

```python
data["Coastal population"].skew()
```

#### Log Transformation
Log transformation was applied to reduce skewness and stabilize variance.

```python

data["log_coastal_population"] = np.log1p(data["Coastal population"])
data["log_total_mismanaged_waste"] = np.log1p(data["Total mismanaged plastic waste in 2010"])
```

```python
data["log_total_mismanaged_waste"].skew()
```

```python
data["log_coastal_population"].skew()
```

The decision to retain or remove log transformation will depend on the assumptions and requirements of the chosen model.

#### Log-Log Relationship Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.scatterplot(
    x=data["log_coastal_population"],
    y=data["log_total_mismanaged_waste"]
)
plt.xlabel("Log Coastal Population")
plt.ylabel("Log Total Mismanaged Waste")
plt.title("Log-Log Relationship")
plt.show()
```
